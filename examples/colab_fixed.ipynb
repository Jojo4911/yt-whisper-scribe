````python
# üöÄ CONFIGURATION AUTOMATIQUE CORRIG√âE
# Cette cellule t√©l√©charge et ex√©cute la version corrig√©e sans erreur

import subprocess
import os

print("üî• YT-WHISPER-SCRIBE - SETUP CORRIG√â")
print("=" * 50)

# T√©l√©chargement de la version corrig√©e
if not os.path.exists('/content/colab_setup_fixed.py'):
    print("üì• T√©l√©chargement du configurateur corrig√©...")
    try:
        # Essai avec la version GitHub corrig√©e
        subprocess.run([
            'wget', '-q', 
            'https://raw.githubusercontent.com/Jojo4911/yt-whisper-scribe/main/examples/colab_setup_fixed.py',
            '-O', '/content/colab_setup_fixed.py'
        ], check=True)
        print("‚úÖ Version corrig√©e t√©l√©charg√©e")
    except:
        # Fallback : cr√©ation manuelle du configurateur
        print("üîß Cr√©ation du configurateur local...")
        
        # Configuration manuelle directe
        !git clone https://github.com/Jojo4911/yt-whisper-scribe.git /content/yt-whisper-scribe
        %cd /content/yt-whisper-scribe
        !pip install -r requirements.txt --quiet
        !apt-get update -qq && apt-get install -y ffmpeg
        
        print("‚úÖ Configuration manuelle termin√©e")
        
        # D√©finition des fonctions essentielles
        def transcribe_video(url, model="turbo", output_format="srt", language="en"):
            import subprocess
            cmd = ["python", "scripts/transcribe.py", url, "--model", model, "--output_format", output_format, "--language", language, "--verbose"]
            result = subprocess.run(cmd)
            if result.returncode == 0:
                from pathlib import Path
                data_dir = Path("data")
                files = list(data_dir.glob(f"*.{output_format}"))
                if files:
                    latest = max(files, key=lambda p: p.stat().st_mtime)
                    print(f"üìÅ Fichier g√©n√©r√©: {latest}")
                    return str(latest)
            return None
        
        print("\nüéâ CONFIGURATION TERMIN√âE (mode manuel)!")
        print("üìñ Utilisez: transcribe_video('URL_YOUTUBE')")

# Import et ex√©cution de la version corrig√©e si t√©l√©charg√©e
if os.path.exists('/content/colab_setup_fixed.py'):
    print("üì§ Chargement du configurateur...")
    exec(open('/content/colab_setup_fixed.py').read())
    print("\nüéâ CONFIGURATION TERMIN√âE!")
    print("üìñ Utilisez: transcribe_video('URL_YOUTUBE')")
else:
    print("‚úÖ Configuration manuelle active")
````

````python
# üîç V√âRIFICATION DU SYST√àME

import os
import subprocess
from pathlib import Path

print("üîç DIAGNOSTIC DU SYST√àME")
print("=" * 40)

# V√©rification GPU
try:
    import torch
    if torch.cuda.is_available():
        gpu_name = torch.cuda.get_device_name(0)
        print(f"‚úÖ GPU: {gpu_name}")
    else:
        print("‚ö†Ô∏è  GPU: Non disponible")
except:
    print("‚ùå PyTorch: Non install√©")

# V√©rification ffmpeg
try:
    result = subprocess.run(["ffmpeg", "-version"], capture_output=True)
    if result.returncode == 0:
        print("‚úÖ ffmpeg: Install√©")
    else:
        print("‚ùå ffmpeg: Erreur")
except:
    print("‚ùå ffmpeg: Non trouv√©")

# V√©rification projet
project_path = "/content/yt-whisper-scribe"
if os.path.exists(project_path):
    print(f"‚úÖ Projet: {project_path}")
    
    # Fichiers importants
    important_files = [
        "scripts/transcribe.py",
        "SWOOD_Glossary.json", 
        "requirements.txt"
    ]
    
    for file_path in important_files:
        full_path = os.path.join(project_path, file_path)
        if os.path.exists(full_path):
            print(f"  ‚úÖ {file_path}")
        else:
            print(f"  ‚ùå {file_path}")
else:
    print(f"‚ùå Projet: Non trouv√© √† {project_path}")

# V√©rification cookies
cookies_paths = [
    "/content/drive/MyDrive/yt-whisper-private/cookies_youtube.txt",
    "/content/cookies_youtube.txt",
    "/content/cookies.txt"
]

print("\nüç™ COOKIES:")
found_cookies = False
for path in cookies_paths:
    if os.path.exists(path):
        print(f"  ‚úÖ {path}")
        found_cookies = True
        
if not found_cookies:
    print("  ‚ö†Ô∏è  Aucun cookie d√©tect√©")

print("\nüìÅ FICHIERS G√âN√âR√âS:")
data_dir = Path(f"{project_path}/data")
if data_dir.exists():
    files = list(data_dir.glob("*"))
    if files:
        for f in sorted(files, key=os.path.getmtime, reverse=True)[:5]:
            size = f.stat().st_size / 1024
            print(f"  üìÑ {f.name} ({size:.1f} KB)")
    else:
        print("  ‚ö†Ô∏è  Aucun fichier g√©n√©r√©")
else:
    print("  ‚ùå Dossier data/ non trouv√©")
````
